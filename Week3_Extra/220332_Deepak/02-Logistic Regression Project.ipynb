# Logistic Regression

Now it's time to do a train test split, and train our model!

You'll have the freedom here to choose columns that you want to train on!

**Split the data into training set and testing set using train_test_split**

```python
from sklearn.model_selection import train_test_split
```

```python
topic=pd.get_dummies(ad_data['Ad Topic Line'],drop_first=True)
city=pd.get_dummies(ad_data['City'],drop_first=True)
country=pd.get_dummies(ad_data['Country'],drop_first=True)
time=pd.get_dummies(ad_data['Timestamp'],drop_first=True)
X=ad_data.copy()
X.drop(['Ad Topic Line','City','Country','Timestamp','Clicked on Ad'],axis=1,inplace=True)
X=pd.concat([X,topic,city,country,time],axis=1)
X.head()
```

```python
y=ad_data['Clicked on Ad']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

**Train and fit a logistic regression model on the training set.**

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression(max_iter=1000)
```

```python
logmodel.fit(X_train,y_train)
```

## Predictions and Evaluations
**Now predict values for the testing data.**

```python
preds=logmodel.predict(X_test)
```

**Create a classification report for the model.**

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,preds))
```
